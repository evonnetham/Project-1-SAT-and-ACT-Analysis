````python
fig = plt.figure(figsize=(15,10))
mask = np.triu(np.ones_like(final.corr()))

cmap = sns.diverging_palette(220, 10, as_cmap=True)

ax = sns.heatmap(final.corr(), 
                 annot = True, 
                 annot_kws = {'size':10}, 
                 mask = mask,
                 cmap = cmap)
 
ax.set_title('Correlation of SAT and ACT', fontsize=20)
y_min, y_max = ax.get_ylim() 
x_min, x_max = ax.get_xlim()
ax.set_ylim(top=y_max+1) 
ax.set_xlim(right=x_max-1);
````

##### Observation 
- Scores of the same test are strongly positively correlated (close to or =1), i.e., ACT english, math, reading, science, composite. And similarly for SAT erw, math and total.
- ACT composite and SAT total scores are weakly negatively correlated (~-0.4) i.e., states with higher ACT score mostly have lower SAT score
- Participation rates are strongly correlation in 2017 and 2018 for both SAT and ACT of about 0.9. However ACT and SAT are strongly negatively correlated with one another 
- Furthermore, total score/composite is negatively correlated to participation rate in both SAT and ACT.

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(df, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2)) 
    fig, ax = plt.subplots(nrows=nrows, 
                           ncols=2, 
                           figsize=(15, 5*nrows), 
                           sharey='all', 
                           sharex='all')
    ax = ax.ravel()
    
    if len(list_of_columns) %2 !=0:
        ax[len(list_of_columns)].set_visible(False)
        
    for i, column in enumerate(list_of_columns): # Gives us an index value to get into all our lists
        ax[i].hist(df[column], bins=15, color='palevioletred') # feel free to add more settings
        ax[i].set_title(list_of_titles[i], fontsize=20)
        ax[i].set_ylabel('State Count', fontsize=15)
        ax[i].set_xlabel(list_of_xlabels[i], fontsize=15)

    plt.tight_layout()
````

Plot the two variables against each other using matplotlib or Seaborn

**Feel free to write a custom function, and subplot if you'd like.** Functions save both time and space.


````python
def subplot_scatter(dataframe, list_of_pairs, list_of_pair_labels, list_of_ax_labels, supertitle):
    nrows = int(np.ceil(len(list_of_pairs) / 2)) # Makes sure you have enough rows   
    fig, ax = plt.subplots(nrows = nrows, ncols = 2, figsize = (12, nrows * 5)) # You'll want to specify your figsize
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for i, pair in enumerate(list_of_pairs): # Gives us an index value to get into all our lists
        alpha_val = 1
        color_val = 'palevioletred'
        sns.regplot(pair[0], pair[1], data = dataframe, ax = ax[i], line_kws = {'color': color_val},
                    scatter_kws = {'color': color_val, 'alpha': alpha_val}) #regplot adds regression automatically
        ax[i].set_title(list_of_ax_labels[i], fontsize = 14) # Do these for all vals
        ax[i].set_xlabel(list_of_pair_labels[i][0])
        ax[i].set_ylabel(list_of_pair_labels[i][1])
    
    plt.suptitle(supertitle, fontsize = 18)
    if len(list_of_pairs) % 2: # delete if it's odd number of pairs to remove an empty plot
        plt.delaxes(ax[len(list_of_pairs)])
    return fig, ax # Return the figure and axes object, to be printed as called
````

````python
pair_list = [('act_math_17', 'sat_math_17'),
             ('act_read_17', 'sat_erw_17'),
             ('act_composite_17', 'sat_total_17'),
             ('sat_total_17', 'sat_total_18'),
             ('act_composite_17', 'act_composite_18')]

pair_label_list = [('ACT Math 2017', 'SAT Math 2017'),
                   ('ACT Reading Score 2017', 'SAT ERW 2017'),
                   ('ACT Composite 2017', 'SAT Total 2017'),
                   ('SAT Total 2017', 'SAT Total 2018'),
                   ('ACT Composite 2017', 'ACT Composite 2018')]

ax_title_list = ['ACT vs. SAT Math (2017)', 
                 'ACT Reading Score vs. SAT ERW (2017)',
                 'ACT Composite vs. SAT Total (2017)',
                 'SAT Total 2017 vs. 2018',
                 'ACT Composite 2017 vs. 2018']

fig, ax = subplot_scatter(final, pair_list, pair_label_list, ax_title_list, 'ACT & SAT 2017/2018 Scatterplots')
````

##### Observation: 

The scatter plots above show that there is a negative correlation between SAT and ACT scores for Math, Reading and Composite/Total scores in both 2017 and 2018. This is because a state which has high SAT participation, thereby a lower mean total score, usually has low ACT participation, thereby a high mean composite score. 

The scatter plots above also show a strong positive correlation for the same exam across multiple years. It indicates that the performance for an individual state is likely to remain consistent for the same exam unless there is a drastic shift in participation rates from one test to another.

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).
